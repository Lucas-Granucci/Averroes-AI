# Objective I Setup

This notebook initializes the directory structure for Objective I data processing, creating folders for metadata, PDFs, extracted text, sentences, API queries, parallel data, and results.

## Setup

### Imports

```python
import yaml
from pathlib import Path
```

### Configuration and Paths

```python
# Load configuration
with open("../config.yaml") as f:
    config = yaml.safe_load(f)

# Set up project paths
DATA_DIR = Path.cwd().parent / "data"
METADATA_DIR = DATA_DIR / "metadata"
PDFS_DIR = DATA_DIR / "pdfs"
EXTRACTED_DIR = DATA_DIR / "extracted"
SENTENCES_DIR = DATA_DIR / "sentences"
API_QUERIES_DIR = DATA_DIR / "api_queries"
PARALLEL_DATA_DIR = DATA_DIR / "parallel_data"
RESULTS_DIR = Path.cwd().parent / "results"
```

## Create Directory Structure

```python
for dir_path in [
    DATA_DIR,
    METADATA_DIR,
    PDFS_DIR,
    EXTRACTED_DIR,
    SENTENCES_DIR,
    API_QUERIES_DIR,
    PARALLEL_DATA_DIR,
    RESULTS_DIR,
]:
    dir_path.mkdir(parents=True, exist_ok=True)

print("Directory structure created successfully!")
```
